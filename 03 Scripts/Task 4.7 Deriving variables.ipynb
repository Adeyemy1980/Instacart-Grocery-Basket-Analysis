```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].max()
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## Create busiest_days column

```python
# Creating frequency for orders day of week
df_ords_prods_merged['orders_day_of week'].value_counts(dropna=False)
```

# The frequency table shows that, the two busiest days are 0 and 1 which is saturday and Sunday while slowest are 3 and 4 which is Wednesday and Tuesday 

```python
days = []

for day in df_ords_prods_merged['orders_day_of week']:
  if day == 0 or day ==1:
    days.append("Busiest days")
  elif day == 3 or day ==4:
    days.append("Lowest days")
  else:
    days.append("Regular days")
```

```python
days
```

```python
df_ords_prods_merged['Busiest days']=days
```

```python
#Frequency of the busiest days
df_ords_prods_merged['Busiest days'].value_counts(dropna=False)
```

# busiest_period_of_day

```python
#Checking frequency for hours of the day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

In a 24hrs, over 2million orders happen between hour of (10, 11,14,15,13,12,16,9,17). Over average  orders happen between(8,18,19,20,7,21,22,23) while Few orders happen (6,0,1,5,2,4,3)

```python
hours = []

for hour in df_ords_prods_merged['order_hour_of_day']:
  if hour in [10, 11,14,15,13,12,16,9,17]:
    hours.append("Most orders")
  elif hour in[8,18,19,20,7,21,22,23]:
    hours.append("Average orders")
  else:
    hours.append("Fewest orders")
```

```python
hours
```

```python
df_ords_prods_merged['busiest_period_of_day']=hours
```

```python
# Checking frequency of the busiest_period_of_day
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna=False)
```
